```python
xn = 512
yn = 512

x = np.linspace(0,1,xn).tolist()
y = np.linspace(0,1,yn).tolist()

x,y = np.meshgrid(x,y)

dx = x[0,1]-x[0,0]
dy = y[1,0]-y[0,0]

a = 1/np.pi*np.sin(np.pi*x)*np.sin(np.pi*y)
v = -np.cos(np.pi*x)*np.sin(np.pi*y)
u =  np.sin(np.pi*x)*np.cos(np.pi*y)

u[:,0]  = -u[:,1]
u[:,-1] = -u[:,-2]
u[0,:]  = u[1,:]
u[-1,:] = u[-2,:]

v[0,:]  = -v[1,:]
v[-1,:] = -v[-2,:]
v[:,0]  = v[:,1]
v[:,-1] = v[:,-2]

u[0,0] = 0
u[0,-1] = 0
u[-1,0] = 0
u[-1,-1] = 0

v[0,0] = 0
v[0,-1] = 0
v[-1,0] = 0
v[-1,-1] = 0
```

```python
device = torch.device("mps")
epochs = 20000
loss_fn = torch.nn.L1Loss()

errors = {}
for layers in [3]:
    for c_h in [16]:
        for pad in ["replicate"]:
            for k in [3, 5]:
                for conv in ["learned", "normal"]:
                    for act in ["gelu"]:
            
                        print(layers, c_h, pad, k, conv, act)
                        net = Net(device, pad, c_h, layers, k, conv, act).to(device)
                        #torch.compile(net)

                        optimizer = torch.optim.Adam(net.parameters(), lr=1e-2)
                        scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 
                                                                    step_size=2000, 
                                                                    gamma=0.5)

                        def get_lr(optimizer):
                            for param_group in optimizer.param_groups:
                                return param_group['lr']

                        running_loss = 0
                        xin = torch.tensor(x).view(1,1,xn,yn).float().to(device)
                        yin = torch.tensor(y).view(1,1,xn,yn).float().to(device)
                        uin = torch.tensor(u).view(1,1,xn,yn).float().to(device)
                        vin = torch.tensor(v).view(1,1,xn,yn).float().to(device)

                        inputs = torch.zeros((1,4,xn,yn)).float().to(device)
                        for ci, c in enumerate([1]): 
                            #1e-3, 5e-3, 1e-2, 5e-2, 
                            #                    1e-1, 5e-1, 1e+0, 5e+0]):

                            inputs[ci,...] = torch.cat((xin,yin,uin*c,vin*c), axis=1)

                        for i in range(epochs):
                            t0 = time.time()
                            optimizer.zero_grad()

                            # Make predictions for this batch
                            u_pred, v_pred, mass_pred, mass_true = net(inputs)

                            loss_mass = torch.mean(torch.abs(mass_pred))
                            loss_mass_true = torch.mean(torch.abs(mass_true))

                            weights = torch.zeros_like(u_pred) + 1 #0.
                            weights[:,:,2:-2,2:-2] = 1

                            loss = loss_fn(u_pred*weights, uin*weights
                               ) + loss_fn(v_pred*weights, vin*weights)
                            loss.backward()

                            # Adjust learning weights
                            optimizer.step()

                            # Gather data and report
                            running_loss += loss.item()
                            t1 = time.time()

                            if i%1000==0:
                                print(i, 
                                      running_loss/(i+1), 
                                      loss_mass.item(),
                                      loss_mass_true.item(),
                                      get_lr(optimizer), 
                                      t1-t0
                                     )

                            scheduler.step() #running_loss/(i+1)

                        for i in range(1):
                            make_plot(inputs[i,2,...].detach().cpu().numpy(), 
                                      u_pred[i:i+1,...].detach().cpu().numpy(), 
                                      inputs[i,3,...].detach().cpu().numpy(), 
                                      v_pred[i:i+1,...].detach().cpu().numpy(),
                                      x, 
                                      y)

                        errors[str(layers) + "_" 
                             + str(c_h) + "_" + pad 
                             + "_" + str(k) + conv 
                             + "_" + act] = running_loss/(i+1)


                        print()
                        print("############################################")
                        print("############################################")
                        print("############################################")
                        print("############################################")
                        print()
```
