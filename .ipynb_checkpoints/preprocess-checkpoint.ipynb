```python
%load_ext autoreload
%autoreload 2

import glob, os, sys

import numpy as np
import matplotlib.pyplot as plt

import torch
from torch.utils.data import TensorDataset, DataLoader
from tabulate import tabulate
from pytorch_networks_convae import *
import argparse
from datasetio import *

import time
from joblib import Parallel, delayed
```

```python
data_dir = "/plp_scr1/agar_sh/data/TPH/"
```

```python
sims = torch.load(data_dir + "/sims.pt")
scale = False
ignore_inds = [8, 39]

def run(an, si):
    sim = sims[si]
    if an == "train":
        check = (sim[1] == "train") 
    else:
        check = (sim[1] == "cv") 
        
    ignr, ignr, raq, fkt, fkp, gr, ar, ignr = sim
    py_dir = data_dir + "/" + sim[1] + "/sim_" + str(sim[0])
    times = torch.load(py_dir + "/times.pt")
    take_every = 1
    
    if 1==1: #not os.path.isfile(py_dir + "/e" + str(take_every) + "_i_vec_select.pt"):
        if check and si not in ignore_inds and len(times)>1:# and (si==0 or si==110):
            print(tabulate([["num", "dataset", "raq", "fkt", "fkp", "gr", "ar"],
                            sim[:-1]
                           ]))

            u  = torch.load(py_dir + "/e" + str(take_every) + "_uprev_data.pt")
            if scale:
                u = scale_var(u, raq, fkt, fkp, "uprev")
            v  = torch.load(py_dir + "/e" + str(take_every) + "_vprev_data.pt")
            if scale:
                v  = scale_var(v, raq, fkt, fkp, "vprev")
    
            p  = torch.load(py_dir + "/e" + str(take_every) + "_pprev_data.pt")
            Tprev = torch.load(py_dir + "/e" + str(take_every) + "_Tprev_data.pt")
            
            times = torch.load(py_dir + "/times.pt")[:-2]
    
            print(len(times), u.shape)
    
            if len(times) > 200:
                rest_steps = np.arange(200, len(times)).tolist()
                
            if len(times)> 1000:
                rest_steps = random.choices(rest_steps, k=min(800,rest_steps[-1]-200))
                i_vec = np.arange(1,200).tolist() + rest_steps
            else:
                i_vec = np.arange(1,len(times)).tolist()

            labels = ["_init.pt", ".pt",]
            
            for label_ind, i_vec in enumerate([i_vec[:5], i_vec[5:]]):
    
                torch.save(u[i_vec,...], py_dir + "/e" + str(take_every) + "_uprev_data_select" + labels[label_ind])
                torch.save(v[i_vec,...], py_dir + "/e" + str(take_every) + "_vprev_data_select" + labels[label_ind])
                torch.save(p[i_vec,...], py_dir + "/e" + str(take_every) + "_pprev_data_select" + labels[label_ind])
                torch.save(Tprev[i_vec,...], py_dir + "/e" + str(take_every) + "_Tprev_data_select" + labels[label_ind])
                torch.save(i_vec, py_dir + "/e" + str(take_every) + "_i_vec_select" + labels[label_ind])
```

```python
results = Parallel(n_jobs=8, verbose=1)(delayed(run)(i,j) 
                              for i in ["train", "cv"] #, "test"] 
                              for j in range(len(sims))
                             )
```
